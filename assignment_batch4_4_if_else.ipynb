### 12. Check if a year is a leap year.

```python
year = 2009

if (year % 4 == 0) and ((year % 100 != 0) or (year % 400 == 0)):
    print('Leap year ->', year)
else:
    print('Non-Leap year ->', year)
```

### 17. Print the grade of a student (A, B, C, D, F) based on marks.
